```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data.groupby('Brand')['Price'].mean()
average_price_by_Brand = data.groupby('Brand')['Price'].mean()
highest_avg_price = average_price_by_Brand
lowest_avg_price = average_price_by_Brand

print("Average Price by Brand:\n", average_price_by_Brand)
print("\nBrand with highest average price:", highest_avg_price)
print("Brand with lowest average price:", lowest_avg_price)
print("Average Price by Brand:\n", average_price_by_Brand)
print("\nBrand with highest average price:", highest_avg_price)
print("Brand with lowest average price:", lowest_avg_price)
```

```python
data.groupby('Status')['Price'].mean()
average_price_by_status = data.groupby('Status')['Price'].mean()
print("Average Price by Condition (Status):\n", average_price_by_status)
```
