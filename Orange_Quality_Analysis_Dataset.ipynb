```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/Orange Quality Data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
corr = data['Size (cm)'].corr(data['Weight (g)'])
print("Correlation between Size and Weight: {correlation}".format(correlation=corr))
```

```python
correlation = data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
print("Correlation between Brix (Sweetness) and pH (Acidity):", correlation)
```

```python

correlation = data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
print("Correlation between Brix (Sweetness) and pH (Acidity):", correlation)
plt.figure(figsize=(10,6))
sns.barplot(x='Brix (Sweetness)',y='pH (Acidity)',data=data)
plt.title("Correlation between Brix (Sweetness) and pH (Acidity)")
plt.xlabel("Brix (Sweetness)")
plt.ylabel("pH (Acidity)")
plt.show()
```

```python
correlation = data['Softness (1-5)'].corr(data['Ripeness (1-5)'])
print(f"Correlation between Softness and Ripeness: {correlation}")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Softness (1-5)', y='Ripeness (1-5)', data=data)
plt.title('Relationship between Softness and Ripeness')
plt.xlabel('Softness (1-5)')
plt.ylabel('Ripeness (1-5)')
plt.show()
```

```python
for variable in ['Size (cm)', 'Weight (g)', 'Brix (Sweetness)', 'pH (Acidity)', 'Ripeness (1-5)']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Variety', y=variable, data=data)
    plt.title(f'{variable} by Variety')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
blemished_softness = data[data['Blemishes (Y/N)'] == 1]['Softness (1-5)'].mean()
unblemished_softness = data[data['Blemishes (Y/N)'] == 0]['Softness (1-5)'].mean()
blemished_ripeness = data[data['Blemishes (Y/N)'] == 1]['Ripeness (1-5)'].mean()
unblemished_ripeness = data[data['Blemishes (Y/N)'] == 0]['Ripeness (1-5)'].mean()
print(f"Average ripeness for fruits with blemishes: {blemished_ripeness:.2f}")
print(f"Average ripeness for fruits without blemishes: {unblemished_ripeness:.2f}")
print(f"Average softness for fruits with blemishes: {blemished_softness:.2f}")
print(f"Average softness for fruits without blemishes: {unblemished_softness:.2f}")
```
